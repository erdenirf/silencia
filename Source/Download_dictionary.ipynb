```python
with open('json_data.json') as json_file:
    DICT_YOUTUBE = json.load(json_file)
len(DICT_YOUTUBE)
```

```python
for key, value in DICT_YOUTUBE.items():
    
    file_exists = os.path.exists('{}.mp4'.format(key))
    if not file_exists and value:
        URL = 'https://www.youtube.com/watch?v=' + value
        yt = YouTube(URL)
        try:
            stream = yt.streams.get_by_itag(18)  #mp4 360p
            stream.download()
        except VideoUnavailable:
            pass
```
