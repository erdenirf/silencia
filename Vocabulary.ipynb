```python
import gensim
import wget
import sys
import logging
from glob import glob
import re
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
import zipfile
import pymorphy2
from erdeni_nlp import Lemmatization
import os
import json
```

```python
morph = pymorphy2.MorphAnalyzer()
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format("220/model.bin", binary=True)
```

```python
folder_name = "Source"
delim = "\\"
re_extension = "*.mp4"
files_lst = glob(folder_name + delim + re_extension)
```

```python
def get_names_list(lst, delimiter):
    return list(map(lambda x: ".".join(x.split(delimiter)[-1].split(".")[:-1]), lst))

vocabularyFilenames = get_names_list(files_lst, delim)
#len(vocabularyFilenames), vocabularyFilenames
```

```python
vocabulary = {}
not_in_model = set()
stopwords_drops = {}
many_words_drops = []

for word in vocabularyFilenames:
    if len(word.split())>1:
        many_words_drops.append(word)
        continue

    code, lemma = Lemmatization(morph, word, drop_stopword=True, convert_upos=True)
    if not code:
        stopwords_drops[lemma] = word + ".mp4"
        continue
        
    if lemma not in model:
        not_in_model.add(lemma)
    
    vocabulary[lemma] = word + ".mp4"
```

```python
with open('vocabulary.json', "w") as f:
    json.dump(vocabulary, f, ensure_ascii=False)
    
with open('stopwords.json', "w") as f:
    json.dump(stopwords_drops, f, ensure_ascii=False)
```
